```python
import os
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datadirectory = '150330_test_data_at_chem'
def makename(filename):
    return os.path.join(datadirectory,filename)
```

```python
filenames = next(os.walk(datadirectory))[2]
print(len(filenames),'\n',filenames)
```

```python
testdata1 = np.loadtxt(makename(filenames[0]))
```

```python
print(testdata1.shape)
print(testdata1)
```

```python
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
```

```python
fig, ax = plt.subplots()
ax.plot(testdata1[:,0],testdata1[:,1])
ax.set_xlabel('Time (s)')
ax.set_ylabel('Signal (V)')
ax.text(2, 0.95*max(testdata1[:,1]), datadirectory+'\n'+filenames[0], bbox=props);
```
