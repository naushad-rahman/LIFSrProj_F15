```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import signal
from __future__ import division
```

```python
TEST_LENGTH = 100000 #Samples (in test)
SAMPLE_RATE = 1000 #Samples / second
```

```python
v = np.random.normal(0, 1, TEST_LENGTH)
V = np.fft.fft(v)
vf_c = 80.0 #Hz
bin_c = vf_c / SAMPLE_RATE * TEST_LENGTH
f_bin_ratio = TEST_LENGTH / SAMPLE_RATE
v_shape = (bin_c/2) / np.arange(1,TEST_LENGTH+1)
v_shape[0:bin_c] = 1 - np.arange(0,bin_c) / (2*bin_c)

fig, (ax, ax2, ax3, ax4, ax5, ax6) = plt.subplots(6, figsize=(8,16))
ax.plot(v)
ax2.plot(abs(V))
ax3.plot(v_shape)
ax4.plot(abs(V)*v_shape)
ax5.plot(np.log10(np.arange(1,TEST_LENGTH+1)), np.log10(abs(V)*v_shape))
```

```python
v = np.random.normal(0, 1, TEST_LENGTH)
V = np.fft.fft(v)
vf_c = 80.0 #Hz
bin_c = vf_c / SAMPLE_RATE * TEST_LENGTH
f_bin_ratio = TEST_LENGTH / SAMPLE_RATE
v_half = (bin_c/2) / np.arange(1,(TEST_LENGTH/2)+1)
v_half[0:bin_c] = 1 - np.arange(0,bin_c) / (2*bin_c)
v_shape = np.concatenate((v_half, v_half[::-1]))
V_prime = v_shape * V
v_prime = np.real(np.fft.ifft(V_prime))

fig, (ax, ax2, ax3, ax4, ax5, ax6) = plt.subplots(6, figsize=(8,16))
ax.plot(v)
ax2.plot(abs(V))
ax3.plot(v_half)
ax4.plot(abs(V[0:TEST_LENGTH/2])*v_half)
ax5.plot(np.log10(np.arange(1,(TEST_LENGTH/2+1))), np.log10(abs(V[0:TEST_LENGTH/2])*v_half))
ax6.plot(v_prime)
```
